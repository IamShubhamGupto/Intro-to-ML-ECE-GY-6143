```python
def plot_function(Wh1x1, Wh1xb, Wh2x1, Wh2xb, Wo1h1, Wo1h2, Whb):
  # do a forward pass
  zh1 = x1*Wh1x1 + Wh1xb
  uh1 = sigmoid(zh1)
  zh2 = x1*Wh2x1 + Wh2xb
  uh2 = sigmoid(zh2)
  y_function = uh1*Wo1h1 + uh2*Wo1h2 + Whb
  # y_function = forward_pass(x1, Wh1x1, Wh1xb, Wh2x1, Wh2xb, Wo1h1, Wo1h2, Whb)
  fig = plt.figure(figsize=(10,10))
  ax = fig.add_subplot()
  ax.plot(x1, y_function)
  # ax.set_xlim([0.0, 1.0])
  # ax.set_ylim([0.0, 2.0])


interact(plot_function, 
         Wh1x1=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         Wh1xb=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         Wh2x1=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         Wh2xb=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         Wo1h1=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         Wo1h2=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         Whb=widgets.FloatSlider(min=-50, max=50, step=0.01, value=1),
         )
```
